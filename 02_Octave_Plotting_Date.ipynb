```octave
t = [0:0.01:0.98];
y1 = sin(2 * pi * 4 * t);
```

```octave
plot(t, y1)
```

```octave
y2 = cos(2 * pi * 4 * t);
```

```octave
plot(t, y2)
```

```octave
plot(t, y1);
hold on;
plot(t, y2, 'r');  % r in red colour
xlabel('time')
ylabel('value')
legend('sin', 'cos')
title('my plot')
print -dpng 'myplot.png'
```

```octave
% close % close down the plot
```

```octave
figure(1); plot(t, y1)
figure(2); plot(t, y2)
```

```octave
subplot(1,2,1); % Divides plot in a 1x2 grid and access first element
plot(t, y1)
subplot(1,2,2);  % second element
plot(t, y2)
axis([0.5 1 -1 1])  % set xrange and yrange
```
